```python
from sklearn_som.som import SOM
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('./breast_cancer_data/breast-cancer-wisconsin.csv')


wine_labels = df.iloc[:, 0]
wine_data = df.iloc[:, 1:]

print(wine_data.head())
print(wine_labels.head())

# Convert to numpy arrays
wine_labels = wine_labels.values
wine_data = wine_data.values

# Normalize the data
scaler = StandardScaler()
wine_data = scaler.fit_transform(wine_data)
```

```python
wine_som = SOM(3,1, wine_data[0].shape[0])
wine_som.fit(wine_data)
```

```python
predictions = wine_som.predict(wine_data)
```

```python
print(predictions)
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
# Plot the results
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(5,7))
x = wine_data[:,0]
y = wine_data[:,1]
colors = [ 'red', 'green', 'blue']

ax[0].scatter(x, y, c=wine_labels, cmap=ListedColormap(colors))
ax[0].title.set_text('Actual Classes')
ax[1].scatter(x, y, c=predictions, cmap=ListedColormap(colors))
ax[1].title.set_text('SOM Predictions')
plt.show()
```
