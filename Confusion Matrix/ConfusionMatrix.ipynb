```python
#librerias utilizadas solo para el modelo
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.datasets import load_breast_cancer
```

```python
def matriz_confusion(y_true, y_pred, target_names):
    """
    Calcula y muestra la matriz de confusión.

    Parámetros:
    - y_true: Lista de etiquetas reales.
    - y_pred: Lista de etiquetas predichas.
    - target_names: Lista de nombres de las clases.
    """

    TP = sum(1 for true, pred in zip(y_true, y_pred) if true == 1 and pred == 1)  # Verdaderos Positivos
    TN = sum(1 for true, pred in zip(y_true, y_pred) if true == 0 and pred == 0)  # Verdaderos Negativos
    FP = sum(1 for true, pred in zip(y_true, y_pred) if true == 0 and pred == 1)  # Falsos Positivos
    FN = sum(1 for true, pred in zip(y_true, y_pred) if true == 1 and pred == 0)  # Falsos Negativos

    print(f"Clases en el dataset: {target_names}")
    for i, target_name in enumerate(target_names):
        print(f"{i} corresponde a {target_name}")

    print("\n\nMatriz de Confusión:")
    print(f"TP (Benigno TP): {TP}")
    print(f"TN (Maligno TN): {TN}")
    print(f"FP (Maligno predicho como Benigno FP): {FP}")
    print(f"FN (Benigno predicho como maligno FN): {FN}")

    print("\n--------------Matriz de Confusión--------------")
    print(f"              Pred: Benigno  Pred: Maligno")
    print(f"Real: Benigno    {TP}            {FN}")
    print(f"Real: Maligno    {FP}            {TN}")


target_names = ["Maligno", "Benigno"]  # Nombres de las clases
matriz_confusion(y, y_pred, target_names)
```

```python
def calcular_accuracy(y_true, y_pred):
    """
    Calcula la precisión (accuracy) del modelo.
    """

    correctos = sum(1 for true, pred in zip(y_true, y_pred) if true == pred)
    total = len(y_true)

    accuracy = correctos / total if total > 0 else 0  # Evitar división por cero
    return accuracy

def calcular_error(y_true, y_pred):
    """
    Calcula la tasa de error del modelo.
    """

    incorrectos = sum(1 for true, pred in zip(y_true, y_pred) if true != pred)
    total = len(y_true)

    error = incorrectos / total if total > 0 else 0  # Evitar división por cero
    return error

def calcular_precision(y_true, y_pred):
    """
    Calcula la precisión (precision) del modelo.
    """
    TP = sum(1 for true, pred in zip(y_true, y_pred) if true == 1 and pred == 1)  # Verdaderos Positivos
    FP = sum(1 for true, pred in zip(y_true, y_pred) if true == 0 and pred == 1)  # Falsos Positivos
    precision = TP / (TP + FP) if (TP + FP) > 0 else 0  # Evitar división por cero
    return precision

def calcular_recall(y_true, y_pred):
    """
    Calcula el recall (sensibilidad) del modelo.
    """
    TP = sum(1 for true, pred in zip(y_true, y_pred) if true == 1 and pred == 1)
    FN = sum(1 for true, pred in zip(y_true, y_pred) if true == 1 and pred == 0)
    recall = TP / (TP + FN) if (TP + FN) > 0 else 0
    return recall

def calcular_ppv(y_true, y_pred):
    """
    Calcula el Positive Predictive Value (PPV), que es lo mismo que precision.
    """
    return calcular_precision(y_true, y_pred)  # PPV es equivalente a precision

def calcular_true_positive_rate(y_true, y_pred):
    """
    Calcula la tasa de verdaderos positivos (True Positive Rate).
    """
    TP = sum(1 for true, pred in zip(y_true, y_pred) if true == 1 and pred == 1)  # Verdaderos Positivos
    FN = sum(1 for true, pred in zip(y_true, y_pred) if true == 1 and pred == 0)  # Falsos Negativos
    tpr = TP / (TP + FN) if (TP + FN) > 0 else 0  # Evitar división por cero
    return tpr

def calcular_true_negative_rate(y_true, y_pred):
    """
    Calcula la tasa de verdaderos negativos (True Negative Rate).
    """
    TN = sum(1 for true, pred in zip(y_true, y_pred) if true == 0 and pred == 0)  # Verdaderos Negativos
    FP = sum(1 for true, pred in zip(y_true, y_pred) if true == 0 and pred == 1)  # Falsos Positivos
    tnr = TN / (TN + FP) if (TN + FP) > 0 else 0  # Evitar división por cero
    return tnr

def calcular_false_positive_rate(y_true, y_pred):
    """
    Calcula la tasa de falsos positivos (False Positive Rate).
    """
    FP = sum(1 for true, pred in zip(y_true, y_pred) if true == 0 and pred == 1)  # Falsos Positivos
    TN = sum(1 for true, pred in zip(y_true, y_pred) if true == 0 and pred == 0)  # Verdaderos Negativos
    fpr = FP / (FP + TN) if (FP + TN) > 0 else 0  # Evitar división por cero
    return fpr

def calcular_false_negative_rate(y_true, y_pred):
    """
    Calcula la tasa de falsos negativos (False Negative Rate).
    """
    FN = sum(1 for true, pred in zip(y_true, y_pred) if true == 1 and pred == 0)  # Falsos Negativos
    TP = sum(1 for true, pred in zip(y_true, y_pred) if true == 1 and pred == 1)  # Verdaderos Positivos
    fnr = FN / (FN + TP) if (FN + TP) > 0 else 0  # Evitar división por cero
    return fnr

def calcular_f1_score(y_true, y_pred):
    """
    Calcula el F1-Score del modelo.
    """
    precision = calcular_precision(y_true, y_pred)
    recall = calcular_recall(y_true, y_pred)
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0  # Evitar división por cero
    return f1
```

```python
accuracy = calcular_accuracy(y, y_pred)
error = calcular_error(y, y_pred)
precision = calcular_precision(y, y_pred)
recall = calcular_recall(y, y_pred)
ppv = calcular_ppv(y, y_pred)
tpr = calcular_true_positive_rate(y, y_pred)
tnr = calcular_true_negative_rate(y, y_pred)
fpr = calcular_false_positive_rate(y, y_pred)
fnr = calcular_false_negative_rate(y, y_pred)
f1 = calcular_f1_score(y, y_pred)

# Imprimir los resultados
print("Accuracy:", accuracy)
print("Error:", error)
print("Precision:", precision)
print("Recall:", recall)
print("Positive Predictive Value (PPV):", ppv)
print("True Positive Rate (TPR):", tpr)
print("True Negative Rate (TNR):", tnr)
print("False Positive Rate (FPR):", fpr)
print("False Negative Rate (FNR):", fnr)
print("F1-Score:", f1)
```

```python
from sklearn.metrics import (
    confusion_matrix,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
)
```

```python
# Calcular la matriz de confusión
matriz_confusion = confusion_matrix(y, y_pred)

# Calcular las métricas
accuracy = accuracy_score(y, y_pred)
precision = precision_score(y, y_pred)
recall = recall_score(y, y_pred)
f1 = f1_score(y, y_pred)

# Extraer valores de la matriz de confusión
TN, FP, FN, TP = matriz_confusion.ravel()

true_positive_rate = TP / (TP + FN)
true_negative_rate = TN / (TN + FP)
false_positive_rate = FP / (FP + TN)
false_negative_rate = FN / (TP + FN)

# Mostrar resultados
print("Matriz de Confusión:")
print(matriz_confusion)
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-Score:", f1)
print("Tasa de Verdaderos Positivos (True Positive Rate):", true_positive_rate)
print("Tasa de Verdaderos Negativos (True Negative Rate):", true_negative_rate)
print("Tasa de Falsos Positivos (False Positive Rate):", false_positive_rate)
print("Tasa de Falsos Negativos (False Negative Rate):", false_negative_rate)
```
